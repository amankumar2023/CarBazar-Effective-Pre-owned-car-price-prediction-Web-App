```python
import pandas as pd               # used to load  dataset to pandas dataframe
import numpy as np  
from sklearn import linear_model
```

```python
df=pd.read_csv('E:\CarBazar\Data_Set.csv')   # loading dataset to  pandas DataFrame
```

```python
for i in range(df.shape[0]):
    df.at[i, 'Company'] = df['Name'][i].split()[0]
    
df
```

```python
df.info()
```

DATA PREPROCESSING

a) Dropping Unecessary variables/missing values

```python
#df1=df.dropna()                                     dropping rows with null values
df1 = df.drop(['New_Price','S No.'], axis = 1)    # dropping NEW_PRICE as it has no significance in the model for prediction
df1
```

```python
df1.isnull()
df1.isnull().sum().sum()
df1.dropna(inplace=True)
df1.info()
```

```python
var = 'Location'
Location = df1[[var]]
Location = pd.get_dummies(Location,drop_first=True)
Location.head()
```

```python
var = 'Fuel_Type'
Fuel_Type = df1[[var]]
Fuel_Type = pd.get_dummies(Fuel_Type,drop_first=True)
Fuel_Type.head()
```

```python
var = 'Transmission'
Transmission = df1[[var]]
Transmission = pd.get_dummies(df1['Transmission'])
Transmission.head()
```

```python
var = 'Owner_Type'
Owner_Type = df1[[var]]
Owner_Type = pd.get_dummies(df1['Owner_Type'])
Owner_Type.head()
```

```python
var = 'Company'
Company = df1[[var]]
Company = pd.get_dummies(df1['Company'])
Company.head()
```

```python
df1=df1.drop(['Name','Fuel_Type','Transmission','Owner_Type','Company'],axis=1) 
df1
```

```python
final_train= pd.concat([df1,Location,Fuel_Type,Transmission,Owner_Type],axis=1)
final_train
```

OUTLIERS

```python
Q1 = final_train.Year.quantile(0.25)
Q3 = final_train.Year.quantile(0.75)
IQR = Q3-Q1
df_final = final_train[~((final_train.Year<(Q1-1.5*IQR)) | (final_train.Year>(Q3+1.5*IQR)))]
df_final
```

```python
Q1 = df_final.Kilometers_Driven.quantile(0.25)
Q3 = df_final.Kilometers_Driven.quantile(0.75)
IQR = Q3-Q1
df_final1 = df_final[~((df_final.Kilometers_Driven<(Q1-1.5*IQR)) | (df_final.Kilometers_Driven>(Q3+1.5*IQR)))]
df_final1
```

```python
Q1 = df_final1.Mileage_kmpl.quantile(0.25)
Q3 = df_final1.Mileage_kmpl.quantile(0.75)
IQR = Q3-Q1
df_final2 = df_final1[~((df_final1.Mileage_kmpl<(Q1-1.5*IQR)) | (df_final1.Mileage_kmpl>(Q3+1.5*IQR)))]
df_final2
```

```python
Q1 = df_final2.Engine_CC.quantile(0.25)
Q3 = df_final2.Engine_CC.quantile(0.75)
IQR = Q3-Q1
df_final3 = df_final2[~((df_final2.Engine_CC<(Q1-1.5*IQR)) | (df_final2.Engine_CC>(Q3+1.5*IQR)))]
df_final3
```

```python
Q1 = df_final3.Power_bhp.quantile(0.25)
Q3 = df_final3.Power_bhp.quantile(0.75)
IQR = Q3-Q1
df_final4 = df_final3[~((df_final3.Power_bhp<(Q1-1.5*IQR)) | (df_final3.Power_bhp>(Q3+1.5*IQR)))]
df_final4
```

```python
model=linear_model.LinearRegression()
```

```python
Y= df_final4.loc[:,['Price']]
Y
```

```python
X= df_final4.loc[:,['Year','Kilometers_Driven','Mileage_kmpl','Engine_CC','Power_bhp','Seats','Location_Bangalore','Location_Chennai','Location_Coimbatore','Location_Delhi','Location_Hyderabad','Location_Jaipur','Location_Kochi','Location_Kolkata','Location_Mumbai','Location_Pune','Fuel_Type_Diesel','Fuel_Type_LPG','Fuel_Type_Petrol','Automatic','Manual','First','Fourth & Above','Second','Third']]
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.1,random_state=0)
model.fit(X_train,Y_train)
```

```python
Y_pred=model.predict(X_test)
score=r2_score(Y_pred,Y_test)
print(score)

print('R2_score: {:2f}%'.format(score*100))
```
